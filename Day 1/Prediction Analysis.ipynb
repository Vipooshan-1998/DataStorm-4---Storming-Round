```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('modified_data_v1.csv')
```

```python
pd.set_option("display.max_columns", None)
```

```python
df.head()
```

```python
df.iloc[100:].head()
```

```python
train = df.iloc[:100].copy()
test = df.iloc[100:].copy()
```

```python
train = train.drop(columns=['shop_id'])
```

```python
test = test.drop(columns=['shop_id'])
```

```python
train['shop_profile'] = train['shop_profile'].replace({'High':2, 'Moderate':1, 'Low':0})
```

```python
train.info()
```

```python
X = train.drop(columns=['shop_profile']).values
Y = train['shop_profile'].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(
    iterations=1000,
    random_seed=0,
    learning_rate=0.01
)
model.fit(
    x_train, y_train,
#     cat_features=cat_cols,
#     eval_set=(X_val, y_val),
#     logging_level='Silent'
)
print('Model is fitted: ' + str(model.is_fitted()))
print('Model params:')
print(model.get_params())
```

```python
pred = model.predict(x_test)
```

```python
prediction = pd.DataFrame()
prediction['Actual'] = y_test
prediction['Presicted'] = pred
prediction
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, pred, average='micro')
```
