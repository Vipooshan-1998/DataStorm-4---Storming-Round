### null values imputation for Train & Test

```python
df['invoice_id'] = df['invoice_id'].fillna(-999)
```

```python
df.info()
```

```python
train = df[~df.item_description.isnull()].copy()
test = df[df.item_description.isnull()].copy()
```

```python
train.shape, test.shape
```

```python
X = train.drop(columns=['item_description']).values
Y = train['item_description'].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
cat_cols = ['customer_id', 'shop_id']
```

```python
from catboost import CatBoostClassifier, Pool

feature_names = list(train.drop(columns=['item_description']).columns)
train_pool = Pool(x_train, y_train, cat_features=cat_cols, feature_names=feature_names)

model = CatBoostClassifier(
    task_type="GPU",
    iterations=1000,
    random_seed=0,
    learning_rate=0.01
)
model.fit(
    train_pool
#     x_train, y_train,
#     cat_features=cat_cols,
#     eval_set=(X_val, y_val),
#     logging_level='Silent'
)
print('Model is fitted: ' + str(model.is_fitted()))
print('Model params:')
print(model.get_params())
```

```python
pred = model.predict(x_test)
prediction = pd.DataFrame()
prediction['Actual'] = y_test
prediction['Presicted'] = pred
prediction
```
